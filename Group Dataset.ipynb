```python
import os
import shutil

# Function to move files listed in the text file to a new existing folder
def move_files(folder_name, filename, destination_folder):
    # Check if the destination folder exists
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder, exist_ok=True)
        # print("Destination folder does not exist.")
        # return

    # Read the list of filenames from the text file
    with open(filename, 'r') as file:
        filenames = file.read().splitlines()
        print(filenames)

    # Move each file to the destination folder
    for filename in filenames:
        # Construct the full path of the .dat and .hea files
        dat_file = os.path.join(folder_name, filename + '.dat')
        hea_file = os.path.join(folder_name, filename + '.hea')

        # Check if the .dat and .hea files exist
        if os.path.exists(dat_file) and os.path.exists(hea_file):
            # Move the .dat file to the destination folder
            shutil.copy(dat_file, destination_folder)
            print(f"Copied '{dat_file}' to '{destination_folder}'.")

            # Move the .hea file to the destination folder
            shutil.copy(hea_file, destination_folder)
            print(f"Copied '{hea_file}' to '{destination_folder}'.")
        else:
            print(f"Files '{dat_file}' or '{hea_file}' not found.")
```

# Bagi dataset into clean, noise, rotated

```python
# Function to move files listed in the text file to a new existing folder
def divide_dataset(folder_name):
    files = os.listdir(folder_name)
    # unique_files = os.basepath(files)

    files.sort()

    num_files = len(files)
    division_1 = num_files // 3
    division_2 = 2 * (num_files // 3)

    # Create target folders if they don't exist
    for folder in ['clean', 'noise']:
        os.makedirs(os.path.join(folder_name, folder), exist_ok=True)

    # Move files to target folders
    for i, file in enumerate(files):
        source_path = os.path.join(folder_name, file)
        if i < division_1:
            destination_path = os.path.join(folder_name, 'clean', file)
        elif i < division_2:
            destination_path = os.path.join(folder_name, 'noise', file)
        else:
            destination_path = os.path.join(folder_name, 'rotated', file)
        shutil.move(source_path, destination_path)
```

```python
import os
import shutil
import random

def move_to_clean(folder_aritmia, folder_normal):
    """Moves half of the randomly selected unique files from 
    folder_aritmia and folder_normal to their respective 'clean' 
    subfolders.
    """

    # Get unique basenames for arrhythmia and normal files
    unique_arrhythmia_files = list({os.path.splitext(f)[0] for f in os.listdir(folder_aritmia)})
    unique_normal_files = list({os.path.splitext(f)[0] for f in os.listdir(folder_normal)})

    # Shuffle both lists
    random.shuffle(unique_arrhythmia_files)
    random.shuffle(unique_normal_files)

    # Get half the number of files
    half_arrhythmia = len(unique_arrhythmia_files) // 2
    half_normal = len(unique_normal_files) // 2

    # Select the first half of each list
    selected_arrhythmia = unique_arrhythmia_files[:half_arrhythmia]
    selected_normal = unique_normal_files[:half_normal]

    # Create destination folders
    ar_clean_folder = os.path.join(folder_aritmia, 'clean')
    # ar_noise_folder = os.path.join(folder_aritmia, 'noise')  # Unused in this function, but you might need it later
    clean_folder = os.path.join(folder_normal, 'clean')
    # noise_folder = os.path.join(folder_normal, 'noise')     # Unused in this function, but you might need it later

    os.makedirs(ar_clean_folder, exist_ok=True)
    os.makedirs(clean_folder, exist_ok=True) 

    # Move arrhythmia files
    for file_base in selected_arrhythmia:
        for ext in [".dat", ".hea"]:
            source_path = os.path.join(folder_aritmia, file_base + ext)
            destination_path = os.path.join(ar_clean_folder, file_base + ext)
            if os.path.exists(source_path):
                shutil.move(source_path, destination_path)

    # Move normal files
    for file_base in selected_normal:
        for ext in [".dat", ".hea"]:
            source_path = os.path.join(folder_normal, file_base + ext)
            destination_path = os.path.join(clean_folder, file_base + ext)
            if os.path.exists(source_path):
                shutil.move(source_path, destination_path)

    print("Done")
```

```python
folder_name_arrhythmia = 'D:/Skripsweet/BISMILLAH_KELAR/input/arrhythmia'
folder_name_normal = 'D:/Skripsweet/BISMILLAH_KELAR/input/normal'
move_to_clean(folder_name_arrhythmia, folder_name_normal)
```
