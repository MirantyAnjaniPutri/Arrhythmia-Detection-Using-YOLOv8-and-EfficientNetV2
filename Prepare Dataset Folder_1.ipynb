# Move the data and hea files

```python
import os
import shutil

# Function to move files listed in the text file to a new existing folder
def move_dathea_files(source_folder, destination_folder):
    # Check if the destination folder exists
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder, exist_ok=True)
        return

    # Iterate through files in the source folder
    for filename in os.listdir(source_folder):
        # Check if the filename ends with .dat or .hea
        if filename.endswith((".dat", ".hea")):
            source_path = os.path.join(source_folder, filename)
            destination_path = os.path.join(destination_folder, filename)

            # Move the file using shutil.move
            shutil.move(source_path, destination_path)
            print(f"Moved '{filename}' to '{destination_folder}'")
```

# Randomize Arrhythmia and Normal to Dataset

```python
import os
import random
import shutil

def move_to_dataset(source, folder_dataset):
    # List all files in the source directory
    files = os.listdir(source)
    
    # Get the base names without extensions and remove duplicates
    base_names = list(set(os.path.splitext(os.path.basename(file))[0] for file in files))

    # Seed the random number generator for reproducibility
    random.seed(42)
    # Shuffle the base names
    random.shuffle(base_names)
    # print(base_names)

    folder_image = os.path.join(folder_dataset, "images")
    folder_label = os.path.join(folder_dataset, "labels")

    # Create the dataset folder if it doesn't exist
    os.makedirs(folder_dataset, exist_ok=True)
    os.makedirs(folder_image, exist_ok=True)
    os.makedirs(folder_label, exist_ok=True)

    # Copy the first 500 files
    for base_name in base_names:
        # Construct the full file paths
        file_png = os.path.join(source, base_name + ".png")
        file_txt = os.path.join(source, base_name + ".txt")
        
        # Construct the destination paths
        dest_png = os.path.join(folder_image, base_name + ".png")
        dest_txt = os.path.join(folder_label, base_name + ".txt")
        
        # Check if the source files exist before copying
        if os.path.exists(file_png):
            shutil.copy(file_png, dest_png)
        if os.path.exists(file_txt):
            shutil.copy(file_txt, dest_txt)
    
    print("Finished moving to dataset")
```

# Divide Train and Test

```python
import os
import shutil
import random

train_images = "D:/Skripsweet/BISMILLAH_KELAR/datasets/images/train"
train_labels = "D:/Skripsweet/BISMILLAH_KELAR/datasets/labels/train"
val_images = "D:/Skripsweet/BISMILLAH_KELAR/datasets/images/val"
val_labels = "D:/Skripsweet/BISMILLAH_KELAR/datasets/labels/val"
# train_labels = "C:/Users/miran/code_skripsi/dataset_skripsi/datasets/labels/train"
# val_images = "C:/Users/miran/code_skripsi/dataset_skripsi/datasets/images/val"
# val_labels = "C:/Users/miran/code_skripsi/dataset_skripsi/datasets/labels/val"

def train_test_split(source, split):
    # List all files in the source directory
    image_folder = os.path.join(source, "images")
    label_folder = os.path.join(source, "labels")
    image_files = os.listdir(image_folder)
    
    # Get the base names without extensions and remove duplicates
    base_names = list(set(os.path.splitext(os.path.basename(file))[0] for file in image_files))

    # Seed the random number generator for reproducibility
    random.seed(42)
    # Shuffle the base names
    random.shuffle(base_names)
    print(base_names)

    # Create the train image folder if it doesn't exist
    os.makedirs(train_images, exist_ok=True)
    # Create the test images if it doesn't exist
    os.makedirs(val_images, exist_ok=True)
    # Create the train label folder if it doesn't exist
    os.makedirs(train_labels, exist_ok=True)
    # Create the test label folder if it doesn't exist
    os.makedirs(val_labels, exist_ok=True)

    test_size = int(len(base_names) * split)

    # Copy the files into train and test folders
    for base_name in base_names[:test_size]:
        # Construct the full file paths
        file_png = os.path.join(image_folder, base_name + ".png")
        file_txt = os.path.join(label_folder, base_name + ".txt")
        
        # Construct the destination paths for test set
        dest_png = os.path.join(val_images, base_name + ".png")
        dest_txt = os.path.join(val_labels, base_name + ".txt")
        
        # Check if the source files exist before copying
        if os.path.exists(file_png):
            shutil.move(file_png, dest_png)
        if os.path.exists(file_txt):
            shutil.move(file_txt, dest_txt)

    for base_name in base_names[test_size:]:
        # Construct the full file paths
        file_png = os.path.join(image_folder, base_name + ".png")
        file_txt = os.path.join(label_folder, base_name + ".txt")
        
        # Construct the destination paths for train set
        dest_png = os.path.join(train_images, base_name + ".png")
        dest_txt = os.path.join(train_labels, base_name + ".txt")
        
        # Check if the source files exist before copying
        if os.path.exists(file_png):
            shutil.move(file_png, dest_png)
        if os.path.exists(file_txt):
            shutil.move(file_txt, dest_txt)
```

```python
# Example usage
split_ratio = 0.2  # 20% for testing, 80% for training
train_test_split("D:/Skripsweet/BISMILLAH_KELAR/datasets", split_ratio)
```

```python
import os
import cv2

def canny_images(file_path):
    image = cv2.imread(file_path)
    canny = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Define the range for blue color (you may need to adjust these values)
    lower_blue = np.array([0, 0, 0])
    upper_blue = np.array([0, 0, 200])
    # Threshold the HSV image to get only blue areas
    mask = cv2.inRange(canny, lower_blue, upper_blue)

    edges = cv2.Canny(mask, 300, 300)

    kernel = np.ones((3, 3), np.uint8)
    dilated_edges = cv2.dilate(edges, kernel, iterations=1)

    # Save the resulting image, overwriting the original image
    cv2.imwrite(file_path, dilated_edges)

train_folder = "C:/Users/miran/code_skripsi/dataset_binarize/images/train"
for file in os.listdir(train_folder):
    image_path = os.path.join(train_folder, file)
    canny_images(image_path)
```

```python
test_folder = "C:/Users/miran/code_skripsi/dataset_binarize/images/val"
for file in os.listdir(test_folder):
    image_path_test = os.path.join(test_folder, file)
    canny_images(image_path_test)
```
