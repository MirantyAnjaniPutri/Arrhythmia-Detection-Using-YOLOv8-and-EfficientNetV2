```python
def json_to_yolo(directory, file):
	with open(file) as f:
		data = json.load(f)

	yolo_data = []

	name = os.path.splitext(file)[0]
	name_with_extension = os.path.basename(file)
	name_without_extention = os.path.splitext(name_with_extension)[0]
	
	file_image = [image for image in os.listdir(directory) 
			   if image.endswith((".jpg", ".jpeg", ".png"))
			   and image.startswith(name_without_extention)]
	
	image_path = os.path.join(directory, file_image[0])

	image = Image.open(image_path)
	width, height = image.size

	# Determine label offset based on directory name
	last_folder = os.path.basename(os.path.normpath(directory))
	label_offset = 0 if last_folder.lower() == 'arrhythmia' else 12

	labels = ['I', 'II', 'III', 'aVF', 'aVL', 'aVR', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']

	# # Create a mapping dictionary
	# label_to_num = {label: idx for idx, label in enumerate(labels)}
	# Create a mapping dictionary with offset
	label_to_num = {label: idx + label_offset for idx, label in enumerate(labels)}

	for lead in data['leads']:
		lead_name = lead['lead_name']
		if lead_name in label_to_num:
			numeric_lead_name = label_to_num[lead_name]
		bounding_box = lead['lead_bounding_box']

		# Extract point 1, 2, 3, and 4 from bounding box
		# height - 
		x1, y1 = bounding_box['0'][1], bounding_box['0'][0]
		x2, y2 = bounding_box['1'][1], bounding_box['1'][0]
		x3, y3 = bounding_box['2'][1], bounding_box['2'][0]
		x4, y4 = bounding_box['3'][1], bounding_box['3'][0]

		# # Prepare YOLO training data entry
		# yolo_entry_label = [
		# 	label, x1 / width, (height - y1) / height,
		# 	x2 / width, (height - y2) / height,
		# 	x3 / width, (height - y3) / height,
		# 	x4 / width, (height - y4) / height
		# ]
		# (height - y1) / height
		yolo_entry_lead = [
			numeric_lead_name, x1 / width, y1 / height,
			x2 / width, y2 / height,
			x3 / width, y3 / height,
			x4 / width, y4 / height
		]

		# Append to YOLO data list
		# yolo_data.append(yolo_entry_label)
		yolo_data.append(yolo_entry_lead)

	with open(name + ('.txt'), 'w') as f:
		for entry in yolo_data:
			# f.write(json.dumps(entry) + '\n')
			f.write(' '.join(map(str, entry)) + '\n')
```

```python
import os

def get_image_path(file, folder, image_extensions):
    # Iterate through image extensions
    for ext in image_extensions:
        image_filename = file + ext
        image_path = os.path.join(folder, image_filename)
        if os.path.isfile(image_path):
            return image_path  # Return the first matching image path
    return None  # Return None if no matching image is found

def convert_json_to_yolo(folder, file, label):
    with open(file) as f:
        data = json.load(f)

    yolo_data = []

    for lead in data['leads']:
        lead_name = lead['lead_name']
        bounding_box = lead['lead_bounding_box']

        # Extract bounding box pixel coordinates
        x_min = min(bounding_box['0'][0], bounding_box['3'][0])  # Min X
        y_min = min(bounding_box['0'][1], bounding_box['1'][1])  # Min Y
        x_max = max(bounding_box['1'][0], bounding_box['2'][0])  # Max X
        y_max = max(bounding_box['2'][1], bounding_box['3'][1])  # Max Y

        image_extensions = ['.jpg', '.jpeg', '.png', '.tiff']
        # Get filename without extension
        filename = os.path.splitext(os.path.basename(file))[0]
        image_path = get_image_path(filename, folder, image_extensions)
        if image_path is None:
            print(f"Image not found for file: {file}")
            continue

        # Prepare YOLO training data entry
        yolo_entry = {
            'image_path': image_path,
            'class': [lead_name, label],  # Multi-label format
            'x_center': (x_min + x_max) / 2,
            'y_center': (y_min + y_max) / 2,
            'width': x_max - x_min,
            'height': y_max - y_min
        }

        # Append to YOLO data list
        yolo_data.append(yolo_entry)

    # Write YOLO training data to a file
    output_file_path = os.path.splitext(file)[0] + '.txt'
    with open(output_file_path, 'w') as f:
        for entry in yolo_data:
            f.write(json.dumps(entry) + '\n')

    print(f"Converted {file} to YOLO format. Saved as {output_file_path}")

def convert_folder_json_to_yolo(folder_path, label):
    # Iterate through files in folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.json'):
            json_file_path = os.path.join(folder_path, filename)
            convert_json_to_yolo(folder_path, json_file_path, label)
```
